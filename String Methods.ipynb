# String Methods

```python
# len()
String_length="Haritha"
Length=len(String_length)
print(f"Length is:{ Length}")

# uppercase/lowercase

name="Nani"
convert=name.lower()
convert_name=name.upper()
print(f"uppercase:{convert}")
print(f"lowercase:{convert_name}")

#strip
Language="  Python and Dotnet   "
Strip=Language.strip()
print(f"result is:{Strip}")

#replace

Languages="Dotnet,Python,Java"
Replace=Languages.replace("Java","C#")
print(f"After replace: {Replace}")

#split

sentence="I am Haritha"
Words=sentence.split()
print(f"After splitting: {Words}")

#join

Words=["you","are","good"]
sentence=' '.join(Words)
print(f"After joining: {sentence}")

#capitalize

words="you are looking good"
result=words.capitalize()
print(f"capitalize result: {result}")

#casefold

sentence="Apple is Good for Health."
result1=sentence.casefold()
print(f"casefold result is: {result1}")

#center

name = "Harika"
x = name.center(10)
print(f"result is:{x}")

#endswith

sentence="Apple is Good for Health."
res=sentence.endswith(".")
print(f"sentence ends with: {res}")

#find

name="Jayanthi"
res_1=name.find("i")
print(f"i index value is:{res_1}")

#isalnum

name="hari12"
res_2=name.isalnum()
print(f"result is: {res_2}")

#isalpha

name="Hari"
res_=name.isalpha()
print(f"result is:{res_}")

#isascii

name="Hari123@"
res_3=name.isascii()
print(f"ascii result is:{res_3}")

#isdecimal

number="14"
number1="20.21"
result1=number.isdecimal()
result2=number1.isdecimal()
print(f"result is:{result1}", f"result is:{result2}")

#isdigit

Digits="212023"
result=Digits.isdigit()
print(f"isdigit result is:{result}")

#isidentifier()

word="2_send" 
check=word.isidentifier()
print(f"isidentifier result is:{check}")

#islower()

color="White"
res=color.islower()
print(f"color is in lowercase:{res}")

#isnumeric()

number="123"
res=number.isnumeric()
print(f"Given number is numeric:{res}")

#isprintable()

word="Hello!\n Are you #1?"
result=word.isprintable()
print(f"Given word is:{result}")

#isspace()

space="    "
space1="  a"
result=space.isspace()
result1=space1.isspace()
print(f"If it contains space ,it is- {result},If it contains anything other than space ,it is- {result1}")

#istitle()

title="Data science"  #Datascience-True
res=title.istitle()
print(f"Title is : {res}")

#isupper()

language="PYTHON IS EASY TO LEARN"
res=language.isupper()
print(f"Given sentence is in uppercase: {res}")

```
