```python
import pandas as pd

df=pd.read_csv(r"C:\Users\iat\Downloads\environment_data.csv")

#will give top 10 rows
print(df.head(20))

#will give bottom 10 rows
print(df.tail(10))
```

```python
#give the summary statistics of all statestics columns
print(df.describe())

#gives information about datatypes in dataset
print(df.info())
```

```python
#will get rows and columns
print(df.shape)

#will give only rows
print(df.shape[0])

#will give only colmns
print(df.shape[1])

print(df.columns)
```

```python
#To check the missing values

print(df.isnull().head(10))
print(df.isnull().sum())
```

```python
#slicing and extracting data

#to get one column data
print(df['year'].head(10))

#to get more than one column data
print(df[['year','class']].head(15))

#will get the data at a specific index
print(df[df.index==21])
```

```python
df.head(20)

#will get data at a specific index
print(df.loc[8])
print(df.iloc[8])

#if we use iloc,we will get the data from 15-20 rows(n-1 will apply)
print(df.iloc[15:21])
#if we use loc,we will get the data from 15-21 rows
print(df.loc[15:21])
```

```python
#dealing with missing data

#1.Dropping missing values
data1=df.dropna()
print(data1.isnull().sum())

#2.Replacing missing values(mean,median)
mean_value=df['data_value'].mean()
print(f"mean_value:{mean_value}")

median_value=df['data_value'].median()
print(f"median_value:{median_value}")
df.fillna(median_value)
df.head(10)

#Renaming column
print(df.rename(columns={'year':'YEAR'},inplace=True))
df.head(10)
```
