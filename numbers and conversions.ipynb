```python
#number
a=5
b=12
c=21
d=20
e=10

print(type(a))
print(type(b))
print(type(c))
print(type(d))
print(type(e))
#float

x= 1.33
y=4.02
z=5.23
X=7.08
H=44.52

print(type(x))
print(type(y))
print(type(z))
print(type(X))
print(type(H))

#Complex

a1=6j
b2=2+3j
c2=4j
dd=8j+3
h1=87.2j

print(type(a1))
print(type(b2))
print(type(c2))
print(type(dd))
print(type(h1))
```

```python
#int,float to complex conversion

ab=72
x=43.77
y=4.00

a=complex(ab)
b=complex(x)
c=complex(y)
m=float(ab) #int to float conversion

print(a)
print(b)
print(c)
print(m)
```
