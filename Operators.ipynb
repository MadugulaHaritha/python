```python
#Assignment Operators

a=5           #Assignment Operator
b=20
c=21
print(a)
print(b)
print(c)

print('---------')

i=4
j=21.9
k=34
i+=7         #Addition Assignment
j+=i
k+=i
print(i)
print(j)
print(k)


print('---------')

d=34
e=4.4
f=67
d-=a   #d=d-a      #Subtraction Assignment
f-=e
e-=20
print(d)
print(f)
print(e)

print('---------')

n=12
b*=e          #Multiplication Assignment
f*=a
n*=a
print(b)
print(f)
print(n)

print('---------')

e/=a         #Division Assignment
a/=n
print(e)
print(a)

print('---------')

n%=5         #Remainder Assignment
s=34
s%=12
print(n)
print(s)

print('---------')

h=7                 #Exponent Assignment
g=3
h**=2
g**=4
print(h)
print(g)
```

```python
# Logical Operators

x1= 23
x2=78
x3=99
x4=50
x5=23

print(x1>x2 and x2<x1)  #Logical and
print(x2>8 and x2>x1)
print(x4>x3 and x1>5)
print(x5>=x1 and x2<x3)
print(x4<=x1 and x1==x5)

print('--------')

print(x4>x3 or x2>x1)   #Logical or
print(x5>x1 or x5==x1)
print(x3<x2 or x2>x3)
print(x5>=x1 or x3<x4)
print(x1==x1 or x5>x4)

print('--------')

print(not(x4>x3 or x2>x1))  #Logical not
print(not(x1==x1 or x5>x4))
print(not(x4<=x1 and x1==x5))
print(not(x1>x2))
print(not(x1==56))
```

```python
#Identity Operators

a1= 87
b1=93
c1=54
d1=123

print(a1 is b1-6)   #is
print(a1 is b1)
print(d1 is a1+10)
print(c1 is b1%a1)
print(d1+68 is c1*3+29)

print('--------')

print(d1 is not b1)   #is not
print(a1 is not b1, d1 is c1) 
print(a1 is not b1-6)
print(a1-6 is not b1)
print(d1 is not d1)
```

```python
#Membership Operators
i= 'Hi','Hello'
j=[14,20,33,4,95]
k='a','h','H','N'

print('H' in i)                #in
print('N' in k and 'Hi' in i)
print(20 in k or 20 in j)
print('hello' in i)
print('r' in i)

print('-------------')

print('Hi' not in i)             #not in
print(33 not in k)
print(20 not in k)
print('hello' not in i)
print(41 not in i and 20 in j)
```

```python
#Bitwise Operators

x=4
y=9
z=14

print(x&y)                #Bitwise and
print(21 & y)
print(y &8 , 45& x)
print(z&x)

print('---------')

print(x|y,y|z)           #Bitwise or
print(34|21)
print(x|y|z)
print(y|23)

print('--------')

print(~x,~y,~z)          #Bitwise not
print(~x&y)
print(~y&x)

print('--------')

print(x^y)                #Bitwise xor
print(6^3)
print(20^21, 12^14)
print(~x^-5)

print('--------')

print(x<<3)            #Left Shift
print(2<<1,21<<3)
print(z<<2)

print('--------')

print(4>>2)           #Right Shift
print(x>>y)
print(z>>x)
```
