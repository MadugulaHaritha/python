```python
import seaborn as sns

ds=sns.get_dataset_names()
print(ds)
```

```python
#Lineplot

data=sns.load_dataset('anagrams')
data
sns.lineplot(x='num1',y='num2',data=data)
```

```python
#Barplot

dataset=sns.load_dataset('anscombe')
dataset
sns.barplot(x='x',y='y',data=dataset)
```

```python
#countplot

data=sns.load_dataset('glue')
data
sns.countplot(x='Score',data=data)
```

```python
#Boxplot
import seaborn 

getdata=sns.load_dataset('planets')
getdata
seaborn.set(style='whitegrid')
sns.boxplot(x='mass',y='distance',data=getdata)
```

```python
#Violin plot

data=sns.load_dataset('tips')
data
seaborn.set(style='whitegrid')
sns.violinplot(x='day',y='tip',data=data)
```

```python
#Stripplot

data=sns.load_dataset('diamonds')
data
sns.stripplot(x='carat',y='price',data=data)
```

```python
#Histplot
data=sns.load_dataset("planets")
data
sns.histplot(data=data,y='year',kde=True)
```

```python
#Pairplot

data=sns.load_dataset("anagrams")
data
sns.pairplot(hue='num1',data=data)
```

```python
#Heatmap
import matplotlib.pyplot as plt
import numpy as np

data=np.random.randint(low=1,high=100,size=(10,10))
cmap = "tab20"
center = 0
annot=True
hm=sns.heatmap(data=data,annot=annot,cmap=cmap,center=center)
plt.show()
```

```python
#Catplot

dt=sns.load_dataset("exercise")
dt
sns.catplot(x="time",y="pulse",hue="kind",data=dt)
```
