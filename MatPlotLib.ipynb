```python
import matplotlib.pyplot as plt
#Line plot

x=[5,8,7]
y=[2,8,9]
plt.plot(x,y)
plt.show()
```

```python
#Bar plot

x=[5,8,7]
y=[2,8,9]
plt.bar(x,y)
plt.show()
```

```python
#Histogram

x=[20,21,40]
plt.hist(x)
plt.show()
```

```python
#Scatter plot

x=[5,8,7]
y=[2,8,9]
plt.scatter(x,y)
plt.show()
```

```python
#Pie chart

label=['Dog','Cat','Cow','Buffalow']
size=[20,21,30,25]
plt.pie(size,labels=label,autopct='%1.1f%%',startangle=45)
plt.show()
```

```python
#Area chat

x=[3,5,7]
y1=[8,6,5]
y2=[8,2,6]
plt.fill_between(x,y1,y2,color="b",alpha=0.4)
plt.plot(x,y1,label="line-1",marker='*')
plt.plot(x,y2,label="line-2",marker='*')
plt.xlabel('x_axis'),plt.ylabel('y_axis')
plt.legend()
plt.show()
```

```python
#Figures

fig=plt.figure(figsize=(2,2),facecolor="Violet",layout="constrained")
fig.suptitle('Figure')
ax=fig.add_subplot()
plt.show()
```

```python
#Sub plots
import numpy as np

#plot 1:
x = np.array([0, 1, 2, 3])
y = np.array([3, 8, 1, 10])

plt.subplot(1, 1)
plt.plot(x,y)

#plot 2:
x = np.array([0, 1, 2, 3])
y = np.array([10, 20, 30, 40])

plt.subplot(2, 1)
plt.plot(x,y)

plt.show()
```
